# Regression - Boston House Prices
Regression is a method to predict a value of a continuous variable. Whereas, Classification is used to predirct discrete and countable variable

`sklearn` is scikit-learn. It is a package with various Machine Learning algorithms. It has various modules within it. One such module is `datasets` which contains various in-built datasets


```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston() # will contain the data, target values along with some metadata
```

```python
print(boston.DESCR) # DESCR stands for description
```

```python
import pandas as pd
```

`boston.data` contains the data (i.e the values of all the 13 attributes) and `boston.target` will hold the target value (i.e. MEDV)

`boston.feature_names` is the data-member which holds the attibute names of the data. (it is a list of strings)

```python
data = pd.DataFrame(boston.data, columns=boston.feature_names) # casting to a dataframe
```

```python
data['MEDV'] = pd.DataFrame(boston.target) # also adding a new coloumn with the target data
```

```python
data.head()
```

Correlation lets us know how the data is related to target. Corelation varies from -1 to +1

```python
pd.DataFrame(data.corr().round(2)) # rounding off covariance to 2 desimal points
```

We're interested with the `MEDV` only. As we need to know how the other variables are related to `MEDV`.

Here we see that the `RM` variable has the highest positive covariance with `MEDV`. Hence, change in `RM` involves a great change in `MEDV`

We need to build the Linear Regression now. As `RM` has the greatest positive covarience, we expect to predict the target (`MEDV`) value accurately using the same.

We are going to build some Linear Regression of the form:

$ MEDV = f(RM) $

Linear Regression is actually that function $ f $ which will let us predict a `MEDV` value for an inputed value of `RM`

It is in the form of:
$ y = f(x) $

As this is a linear regression, the function must be a linear equation (i.e. a polynomial of degree 1 - essentially a straight line)

Hence, it can be:
$
y = mx + c \\
y = w_{1}x + w_{0}
$

In ML literature, these are called weights ($ w_{0} $ and $ w_{1} $). We need to find the weights


```python
x = data['RM']
y = data['MEDV']
```

```python
pd.DataFrame([x, y]).transpose().head() # just visualizing the data
```

To know how good our model is performing, we will have to split the data into training and testing data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
# test_size should be between 0 and 1. 0.3 means 30% of the data will be used for testing

# These are Series type object
print(type(x_train))
print(type(y_train))
```

```python
# target can be Series, but the feature/attributs should be in a DataFrame or ndArray for sklearn linear model. As trget can be only 1 type of data, but features can be multiple
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression() # this is just a blank model now
model.fit(x_train, y_train) # model is trained here

y_pred = model.predict(x_test) # passing all the test cases
```

```python
# We already know the true values. We will get the values that the model predicted in y_pred. So, we can now compare the true values with the predicted value to calculate the amount of error.

from sklearn.metrics import mean_squared_error
from math import sqrt

print(sqrt(mean_squared_error(y_pred, y_test))) # using root mean square to calculate the error
# ideally error should be zero
```

```python
# We are getting the coefficient and intercept that the model has computed. As weare using linear regression, we only have one coefficient

print("MEDV = {}RM + {}".format(model.coef_[0].round(2), model.intercept_.round(2)))
# printing the equation
```

```python
from matplotlib import pyplot as plt
 # magic function for jupter notebook - embedding plotting
%matplotlib inline
```

```python
plt.scatter(x_test, y_test, label="Actual") # scatter plot of actual values
plt.plot(x_test, y_pred, color="red", label="fit") # the linear regression model
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.legend()
plt.show()
```
